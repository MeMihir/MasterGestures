```python
import numpy as np
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('mainDF.csv', index_col=0)
```

```python
enc = LabelEncoder()
enc.fit(df.iloc[:,-1:].values)
class_mapping = enc.classes_
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,:-1].values, df.iloc[:,-1:].values, test_size=0.33, random_state=42)
```

```python
y_test = enc.transform(y_test)
y_test = np.expand_dims(y_test, 1)
y_train = enc.transform(y_train)
y_train = np.expand_dims(y_train, 1)
```

```python
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)
watchlist = [(dtest, 'valid'), (dtrain, 'train')]
param = {'objective':'multi:softmax',
         'eta':'0.3', 'max_depth':10,
         'silent':1, 'nthread':-1,
         'num_class':len(enc.classes_),
         'eval_metric':'merror'}
model = xgb.train(param, dtrain, 50, watchlist, early_stopping_rounds=20,
                  verbose_eval=10)
```

```python
model.save_model('gesture_classifier.json')
```
